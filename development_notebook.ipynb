```julia
component_names = Set()
for component in get_components(Component, sys)
    name = get_name(component)
    println("Component: $name, Type: $(typeof(component))")

    if name in component_names
        println("WARNING: Duplicate component name: $name")
    else
        push!(component_names, name)
    end
end

# Check bus indexing specifically
buses = get_components(ACBus, sys)
println("Bus information:")
for bus in buses
    println("  $(get_name(bus)): number=$(get_number(bus)), base_voltage=$(get_base_voltage(bus))")
end
```
